```python
def actionChange(actions1, actions2, user1, user2, rt):
    if len(actions1) != len(actions2): return "Lenghts mismatch" #Raise exception
    rounds = len(actions1)
    prev1 = ""
    prev2 = ""
    switch_events = pd.DataFrame(columns=["type", "lag", "subject", "round", "rt"])
    lag = 0
    for r in range(0, rounds):
        if r ==0:
            prev1 = actions1[r]
            prev2 = actions2[r]
            lag += 1
            continue
        elif actions1[r] != prev1:
            switch_events = switch_events.append(
                {"context": prev2, "type": "%s%s" % (prev1, actions1[r]),
                 "lag": lag, "subject": user1, "round": r + 1, "rt": rt[r]}, ignore_index=True)
            lag = 0
        if actions2[r] == prev2:
            lag += 1
        else: lag = 1
        prev1 = actions1[r]
        prev2 = actions2[r]
    return switch_events
```

```python
def actionCounts(actions, user):
    prosocial = ['CCC', 'CDC', '00C', 'DDC', 'DCC']
    proself = ['DDD', 'DCD', '00D', 'CCD', 'CDD']
    rating = 0
    act_df = {'user':user,'00D':0,'00C':0,'DCD':0,'CDC':0,'CCC':0,'CDD':0,'CCD':0,'DDC':0,'DCC':0,'DDD':0, 'type':"", 'score': 0}
    for a in actions:
        act_df[a] = act_df[a] + 1
        if a in prosocial:
            rating = rating + 1
        else: rating = rating - 1
    act_df["score"] = rating
    if rating >= 0: act_df["type"] = "Pro-social"
    else: act_df["type"] = "Pro-self"
    return act_df
```

```python
switches = pd.DataFrame()
actions = pd.DataFrame()
for t in data.treatment.unique():
    t_df = data.loc[(data.treatment == t)]
    users = t_df.player.unique()
    switchesT = pd.DataFrame()
    actionsT = pd.DataFrame()
    for u in users:
        df = t_df.loc[t_df.player == u]
        df = df.sort_values(by = "round", ascending = True)
        opp = str(df.opponent.unique())
        acc1 = df.action_player.values
        acc2 = df.action_opponent.values
        rt = df["rt"].values
        switchesT = switchesT.append(actionChange(acc1, acc2, u, opp, rt))
        actionsT = actionsT.append(actionCounts(df.actions, u), ignore_index=True)
    switchesT["treatment"] = t
    actionsT["treatment"] = t
    switches = switches.append(switchesT, ignore_index = True)
    actions = actions.append(actionsT, ignore_index = True)
```

```python
data["prev"] = data['actions'].str.slice(1)
```

```python
dict_other = {
    "0C":0,
    "0D":0,
    "CD":0,
    "CC":3,
    "DC":4,
    "DD":1
}
dict_self = {
    "0C":0,
    "0D":0,
    "CD":4,
    "CC":3,
    "DC":0,
    "DD":1
}
```

```python
data["self"] = 0
data["other"] = 0
for index, row in data.iterrows():
    data.loc[index, "self"] = dict_self[row["prev"]]
    data.loc[index, "other"] = dict_other[row["prev"]]
```

```python
data["self"] = pd.to_numeric(data["self"])
data["other"] = pd.to_numeric(data["other"])

mean_self = data[["player", "self"]].groupby("player").mean()["self"]
mean_other = data[["player", "other"]].groupby("player").mean()["other"]

actions = actions.merge(mean_self, left_on="user", right_on= "player")
actions = actions.merge(mean_other, left_on="user", right_on= "player")
actions
```

```python
import math
actions["radians"] = np.arctan(actions["other"] /actions["self"])
actions["degrees"] = (np.arctan(actions["other"] / actions["self"] )) * 180.0 / math.pi

actions
#altruists: 57.15 < angle
# prosocials: 22.45 < angle < 57.15
# individualists: -12.04 < angle < 22.45
# competitors: angle < -12.04
```

```python
## Adjusting weights for actions
actions["type_adj"] = ""
actions["score_adj"] = 0

for u in actions.user.unique():
    score = sum(actions.loc[actions.user == u].values[0][0:10] * [3,-4,0,-4,4,0,0,-4,4,0])
    actions.loc[actions.user == u, "score_adj"] = score
    actions.loc[actions.user == u, "type_adj"] = "Pro-social" if score>=0 else "Pro-self"
actions
```

```python
#Receives a list of players to show
def get_params(param, players, treatment = "fix", column2 = None):
    param_cols = ["%s%d" % t for t in zip([param] * 10, [x*10 for x in range(1, 10+1)])]
    param_sd = ["%s_sd%d" % t for t in zip([param] * 10, [x*10 for x in range(1, 10+1)])] 
    params = pd.melt(params_fix.loc[params_fix.user.isin(players)], id_vars=['user'], value_vars = param_cols)
    sds = pd.melt(params_fix.loc[params_fix.user.isin(players)], id_vars=['user'], value_vars = param_sd)
    d = params.groupby("variable").mean()
    d.index = pd.to_numeric(d.index.str.replace(param, ""))
    d_sd = sds.groupby("variable").mean()
    d_sd.index = pd.to_numeric(d_sd.index.str.replace("%s_sd" % (param), ""))
    d_sd["round"] = d_sd.index
    d["round"] = d.index
    d = d.merge(d_sd, on = "round")
    d = d.sort_values(by="round")
    d.rename(columns = {"value_x":param, "value_y":"std"}, inplace=True)
    return d
```
