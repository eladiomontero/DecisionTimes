The drift difussion model (DDM) is a type of sequential sampling model (SSM) which, until now, has been uniquely applied in impulsive / non-deliberated two-choice decisions. These SSM models, and thus the DDM, assume that decisions are made by accumulating information until a threshold is reached. In order to model this accumulation of knowledge, DDM uses a random walk in two dimensions, starting at point z and finishing when it reaches a decision threshold (a or 0). 

The starting point z can be understood as an initial bias, which can be due to previous experiences or intuitive notions. On the other hand, the boundaries represent the minimum accumulated knowledge needed to make a decision, so harder decisions will have wider boundaries than simple choices.

The speed by which the information is accumulated is defined as $v$ (drift rate) and it is a constant value throughout the experiment. In past papers, the drift rate $v$ has been defined positive when the accumulation of information led to the right decision (and negative in the opposite case) but, as decisions in cooperation-defection experiments are not right or wrong, $v$ will be defined as positive when the accumulation of knowledge leads to a cooperative choice (and negative on the contrary). 

Apart from the drift rate, there is a source of stochastic noise in the decision process. Using the same parameters v and z, this noise will lead to different final decisions and different final times. Thevariability in decision times will produce the characteristic distribution of RT (reaction time) and the two possible final decisions will bring errors and the corresponding error RT distribution.

One last parameter $t_0$?

The equation which represents the probability distribution:
$$P(t| v,a,z)=\frac{\pi}{a^2}\exp{(-vza-\frac{v^2 t}{2})} Ã— \sum_{k=1}^\infty k \exp{(-\frac{k^2\pi ^2 t}{2a^2})}\sin{(k\pi z)}$$

# File parsing

```python
def parse_files(files_to_parse,outfile):
    o=open(outfile+".dat","w")
    for index,i in enumerate(files_to_parse):
        if index==0:
            for line in open("Esther/history_"+i+".dat"):
                o.write(line.replace(" ",",").replace("usuario",i+"_usuario")) 
        else:
            f=open("Esther/history_"+i+".dat")
            f.readline()
            for line in f:
                o.write(line.replace(" ",",").replace("usuario",i+"_usuario") ) 
    o.close()
    return 
```

# Data preprocessing

```python
def preprocessing(data):#only user (not autom)
    data=data[data["who_played"]=="user"]

    #delete problematic columns
    del data["who_played"], data["opponent"], data["payoff"], data["time_php"], data["action_opponent"]

    #rename rt, userid and response columns
    data.rename(columns={'time_js':'rt', "action_player":"response", "player":"userid"}, inplace=True) 

    #make rt integers and seconds
    data['rt'] = data.rt.astype(int)/1000

    #coop is 1, defect is 0
    data.loc[data['response'] == "C", "response"] = 1.0
    data.loc[data['response'] == "D", "response"] = 0.0

    #standard deviation for bin width selection
    print("STD of data is ", data["rt"].std())

    #rt which are defective become neg
    data = hddm.utils.flip_errors(data)

    #see how many decisions we hav
    print("N of decisions are ",data.shape[0])

    #see how the df looks like
    data.head()
    return data
```

# Times distribution 

```python
def plot_norm_distrib(output_filename,data,logscale):
    fig = plt.figure(figsize=(14,10), )
    if logscale:
        ax = fig.add_subplot(111, xlabel='RT (s)', ylabel='p(t) norm. count', title='RT distributions (LOG SCALE)'+output_filename)
        data.rt.hist(log=True,bins=300,  ax=ax, density=True, histtype="stepfilled")
        plt.legend(["coop: +; defect: -"])
        plt.xlim(-30,30)
        plt.savefig(output_filename+'_logscale.png')
    else:
        ax = fig.add_subplot(111, xlabel='RT (s)', ylabel='p(t) norm. count', title='RT distributions '+output_filename)
        data.rt.hist(bins=300,  ax=ax, density=True, histtype="stepfilled")
        plt.legend(["coop: +; defect: -"])
        plt.xlim(-30,30)
        plt.savefig(output_filename+'.png')
    return
```

# Get graph from parameters

```python
def DDM(t,a,v,t0,z,kmax=1000):
    if t < t0:
        return 0
    else:
        x = t-t0
        c1 = pi/a**2 * exp(-v*z*a - v**2*x/2)
        c2 = 0
        for k in range(1,kmax):
            c2 += k*exp(-k**2*pi**2*x/(2*a**2))*sin(k*pi*z)
        return c1*c2
```

# Bin size
For any histogram, the bin size follows eq (https://www.fmrib.ox.ac.uk/datasets/techrep/tr00mj2/tr00mj2/node24.html):
$$W=3.49 \sigma N^{-1/3}$$

here $\sigma$ equals 2,94 and N equals 17200.
$$W=0.39749341805822797$$
so the number of bins has to be
$$60/W=150$$
as the data is split in 2 sides (positive for cooperation times and negative for defection times) we use 300 bins.

# Quality of data
In order to assess the quality of the data, a statistical test is carried out over the following parameters: drift rate v, initial bias z, non-decision time t and decision boundary a. A first statistical analysis can be done with the autocorrelation, the trace and the histogram for each parameter. 

Autocorrelation coefficient at lag k is obtained from the following eq.
$$r_k=\frac{\sum_{i=1}^{N-k} (x_i-\bar{x})(x_{i+k}-\bar{x})}{\sum_{i=1}^N(x_i-\bar{x})^2}$$

# Model fitting data

```python
def get_model(data,sampleN,burnN):
    model = hddm.HDDM(data, include='z')
    model.sample(sampleN, burn=burnN)
    stat=model.gen_stats() 
    #print(stat["mean"])
    model.plot_posteriors()
    model.plot_posterior_predictive()
    plt.title('Posterior predictive')
    plt.xlabel('RT')
    plt.ylabel('Probability density')
    return stat


def get_manual_fitting(output_filename,data,stat,nbins,ylimit,xlimit):
    val_neg=[DDM(i/100,stat["mean"]["a"],stat["mean"]["v"],stat["mean"]["t"],stat["mean"]["z"]) for i in range(100*xlimit)]
    val_pos=[DDM(i/100,stat["mean"]["a"],-1*stat["mean"]["v"],stat["mean"]["t"],1-stat["mean"]["z"]) for i in range(100*xlimit)]
    xval_neg=[-i/100 for i in range(100*xlimit)]
    xval_pos=[i/100 for i in range(100*xlimit)]

    fig = plt.figure(figsize=(14,10), )
    ax = fig.add_subplot(111, xlabel='RT (s)', ylabel='p(t) norm. count', title='RT distributions ')
    #dataOut.rt.hist(bins=300,  ax=ax, density=True, histtype="stepfilled")
    plt.plot(xval_pos,val_pos)
    plt.plot(xval_neg,val_neg)
    plt.xlim(-xlimit*1.5,xlimit*1.5)
    plt.yscale("log")

    counts,bin_edges = np.histogram(data["rt"],nbins,normed=True)
    bin_centres = (bin_edges[:-1] + bin_edges[1:])/2.
    plt.errorbar(bin_centres, counts, yerr=0, fmt='o')
    plt.ylim(ylimit)
    plt.legend(stat["mean"])
    plt.savefig(output_filename+".png")
    plt.show()
    return
```
